```python
import pydicom
```

READING THE DICOM FILE USING THE DCMREAD

```python
dicom_data = pydicom.dcmread('/content/000000.dcm')
dicom_data
```

ACCESSING ATTRIBUTES IN THE DICOM DATASET

```python
#This accesses the patient name attribute displaying its element value

patient_name = dicom_data.PatientName
patient_name
```

```python
Patient_Age = dicom_data.PatientAge
Patient_Age
```

```python
# Acessing the DICOM attribute (Patient name) using its unique tag
#Note that using the unique tag yields the whole value of the dictionary

dicom_data[0x0010,0x0010]
```

```python
dicom_data[0x0010, 0x1010]
```

```python
#Specifying only the elements value using the key tag

dicom_data[0x0010,0x0010].value
```

```python
#This displays only the keys
dicom_data.keys()
```

```python
# This displays only the values
dicom_data.values()
```
